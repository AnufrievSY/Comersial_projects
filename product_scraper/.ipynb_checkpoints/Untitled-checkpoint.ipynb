# Загружаем необходимые библиотеки:

```python
import time  # для подсчета затрчиваемого времени

import requests  # извлечение данных с сайта
from http.client import responses  # расшифровка кода резултатов откработтки функции requests.get(url)
from bs4 import BeautifulSoup  # извлечение данных из файлов HTML и XML.
from lxml import html  # для извлечения данных из файлов HTML

from IPython.display import clear_output  # отчистка вывода перед принтом новых результатов

import pandas as pd  # создание базы данных
```

```python
# Задаем список магазинов, сведения из которых мы хотим получить
shops_dir = {
    'Пятерочка': 'https://eda.yandex.ru/retail/paterocka?placeSlug=pyaterochka_zypgf',
    'Магнит': 'https://eda.yandex.ru/retail/magnit?placeSlug=magnit_tmvbx',
    'Яндекс Лавка': 'https://lavka.yandex.ru/54',
    'Перекресток': 'https://eda.yandex.ru/retail/perekrestok?placeSlug=perekrestok_qtrvb',
    'Лента гипер': 'https://eda.yandex.ru/retail/lenta?placeSlug=lenta_ytkcx',
    'Лента супер': 'https://eda.yandex.ru/retail/lenta_onlajn?placeSlug=super_eyfpw',
    'Верный': 'https://eda.yandex.ru/retail/lenta_onlajn?placeSlug=super_eyfpw',
    'Окей': 'https://eda.yandex.ru/retail/okej?placeSlug=okej_mrfea',
    'Елисей': 'https://eda.yandex.ru/retail/elisej?placeSlug=elisej_twplf'
}

# Задаем датафрейм в которым будем записывать результаты
df = pd.DataFrame(columns = ['ID', 'Название', 'Цена', 'Вес', 'Описание', 'Подкатегория', 'Категория', 'Магазин'])
```

```python
start = time.time()
mean_time_list = []
# Запускаем цикл прохода по каждому магазину
for m, (shop, URL) in enumerate(shops_dir.items()):
    if len(mean_time_list) == 0:
        time_print = ''
        pass
    else:
        clear_output(wait=True)
        mean_time = sum(mean_time_list)/len(mean_time_list).round(2)
        time_print = f'Среднее время работы составляет: {mean_time} секунд\nПримерное время до конца работы: {mean_time*(len(shops_dir)-m)} секунд'
        print(time_print)
        
    start2 = time.time()
    print(f'{shop}: start')

    # отправляем получаем данные с переданного сайта
    response = requests.get(URL)
    print(f'{shop}: get status {responses[response.status_code]}')

    # получаем сведения из запроса
    tree = html.fromstring(response.text)

    # создаем словарь с названием категорий магазина и ссылкой на категорию
    name_list= tree.xpath('//*[@id="root"]/div/div/div[1]/div/div/div/aside[1]/div[2]/div[2]/ul/li//text()')
    link_list = ['https://eda.yandex.ru'+l for l in tree.xpath('//*[@id="root"]/div/div/div[1]/div/div/div/aside[1]/div[2]/div[2]/ul/li//@href')]
    category_dir = {name_list[i]: link_list[i] for i in range(len(name_list))}
    print(f'{shop}: найдено {len(category_dir)} категорий')

    # запускаем цикл прохода по каждой категории
    for n, (category, url) in enumerate(category_dir.items()):
        clear_output(wait=True)
        print(time_print)
        print(f'{shop}: start')
        print(f'{shop}: get status {responses[response.status_code]}')
        print(f'{shop}: найдено {len(category_dir)} категорий')
        print(f'{shop}: категория {n} из {len(category_dir)}')

        # получаем сведения по опредленной категории магазина
        request = requests.get(url).text
        tree = html.fromstring(request)

        # обозначаем xpath выбранной категории
        path_1_lvl = '/html/body/div[1]/div/div/div[1]/div/div/div/main/div[1]/div'

        # получаем кол-во подкатегорий в данной категории
        count_of_category = len(tree.xpath(path_1_lvl))

        # запускаем цикл прохода по подкатегории
        for i in range(3,count_of_category+1):

            # получаем название подкатегории
            subcategory = tree.xpath(f'{path_1_lvl}[{i}]/div/div/h2/a//text()')[0]

            # обозначаем xpath подкатегории
            path_2_lvl = f'{path_1_lvl}[{i}]/ul/li'

            # получаем кол-во продуктов в подкатегории
            count_of_items = len(tree.xpath(path_2_lvl))          

            # запускаем цикл прохода по каждому продукту
            for j in range(1, count_of_items+1):

                # получаем сведения о продукте
                item = tree.xpath(f'{path_2_lvl}[{j}]//text()')

                # если имеются хоть какие-то сведения то запускаем скрипт, если нет, то пропускаем этот элемент
                if len(item) != 0:
                    try:
                        # если в названии элемента если скидка то получаем финальную цену после скидки, имя и вес
                        if '%' in item[0]:
                            price = item[1].split('\u2009')[0]
                            name = item[2]
                            weight = ' '.join(item[3].split('\xa0'))
                        else:
                            price = item[0].split('\u2009')[0]
                            name = item[1]
                            weight = ' '.join(item[2].split('\xa0'))

                        # получаем описание товара 
                        description = tree.xpath(f'{path_2_lvl}[{j}]/a/div//@description')

                        # делаем запись в датафрейм ID формируем из порядкового номера магазина, №категори, №номера подкатегории, №товара
                        df = pd.concat([pd.DataFrame([[f'{m}{n}{i-2}{j}', name, price, weight, description, subcategory, category, shop]], columns=df.columns), df], ignore_index=True)
                    except:
                        pass
                else:
                    pass

    mean_time_list.append(time.time()-start2) 
    
print(f'Весь скрипт отработан за {time.time()-start} секунд.')
# сохраняем получившийся результат                
df.to_csv('data_base.csv', index = False)
```

```python
def find_item(shop, category, subcategory, find):
    URL = shops_dir[shop]
    response = requests.get(URL)
    tree = html.fromstring(response.text)

    name_list= tree.xpath('//*[@id="root"]/div/div/div[1]/div/div/div/aside[1]/div[2]/div[2]/ul/li//text()')
    link_list = ['https://eda.yandex.ru'+l for l in tree.xpath('//*[@id="root"]/div/div/div[1]/div/div/div/aside[1]/div[2]/div[2]/ul/li//@href')]
    category_dir = {name_list[i]: link_list[i] for i in range(len(name_list))}
    url = category_dir[category]

    request = requests.get(url).text
    tree = html.fromstring(request)
    path_1_lvl = '/html/body/div[1]/div/div/div[1]/div/div/div/main/div[1]/div'
    count_of_category = len(tree.xpath(path_1_lvl))
    number_of_subcategiry = ''
    for i in range(3,count_of_category+1):
        subc = tree.xpath(f'{path_1_lvl}[{i}]/div/div/h2/a//text()')[0]
        if subc == subcategory:
            subcategory = i

    path_2_lvl = f'{path_1_lvl}[{subcategory}]/ul/li'
    count_of_items = len(tree.xpath(path_2_lvl))          
    for j in range(1, count_of_items+1):
        item = tree.xpath(f'{path_2_lvl}[{j}]//text()')
        if find in item:    
            return item
        else:
            pass


find_item("Окей", 
          "Для детей", 
          "Бытовая химия", 
          "Кондиционер концентрат для белья детский ТМ О'кей")
```
