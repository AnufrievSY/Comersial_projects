```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import time  # для подсчета затрчиваемого времени
from IPython.display import clear_output # отчищает вывод
import matplotlib.pyplot as plt # строит график
from concurrent.futures import as_completed
from requests_futures.sessions import FuturesSession
from concurrent.futures import ProcessPoolExecutor

def get_future_soup(url): 
    with FuturesSession() as session:
        session.headers = {
                        'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
                        'Accept-Encoding' : 'gzip, deflate, br',
                        'Accept-Language' : 'ru,en;q=0.9',
                        'Cache-Control' : 'max-age=0',
                        'Connection' : 'keep-alive',
                        'Cookie' : 'currentCourt=; _ym_uid=1675477735944162266; _ym_d=1675477735; _ym_isad=1; c-dis=1',
                        'Host' : 'mos-sud.ru',
                        'Referer' : 'https://mos-sud.ru/search?caseDateFrom=&caseDateTo=&caseFinalDateFrom=&caseFinalDateTo=&caseLegalForceDateFrom=&caseLegalForceDateTo=&caseNumber=&codex=12.8&documentStatus=&documentText=&documentType=&judge=&participant=&processType=3&publishingState=&uid=&year=&formType=fullForm&page=2',
                        'sec-ch-ua' : '"Not?A_Brand";v="8", "Chromium";v="108", "Yandex";v="23"',
                        'sec-ch-ua-mobile' : '?0',
                        'sec-ch-ua-platform' : '"Windows"',
                        'Sec-Fetch-Dest' : 'document',
                        'Sec-Fetch-Mode' : 'navigate',
                        'Sec-Fetch-Site' : 'same-origin',
                        'Sec-Fetch-User' : '?1',
                        'Upgrade-Insecure-Requests' : '1',
                        'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 YaBrowser/23.1.1.1138 Yowser/2.5 Safari/537.36',
                    }

        if type(url) == list:
            future = [session.get(u, headers = {'User-agent': 'your bot 0.1'}, verify=False).result() for u in url]
            soup = [BeautifulSoup(f.text, "html.parser") for f in future]
        else:
            future = session.get(url, headers = {'User-agent': 'your bot 0.1'}, verify=False).result()
            soup = BeautifulSoup(future.text, "html.parser")        
    return soup

def get_soup(url):
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    return soup
```

```python
main_url = 'https://quke.ru/'
soup = get_soup(main_url)
df_category = pd.DataFrame(columns = ['Ссылка', 'Наименование', 'Кол-во товаров'])
r = soup.find_all('div', {'class': 'b-cat-2__col'})
for e in r:
    get_url = e.find('a', {'class':"b-cat-2__card-img-wr"})['href']
    get_category_name = e.find('a', {'class' :"b-cat-2__card-name"}).text
    get_count_in_category = int(e.find('div', {'class':"b-cat-2__quantity"}).text.split(' ')[0])
    df_category = df_category.append({'Ссылка':get_url, 'Наименование':get_category_name, 'Кол-во товаров': get_count_in_category}, ignore_index = True)
df_category
```

```python
def get_info(element, i, j):
    return element.find(i, {'class': f"b-card2-v2__{j}"})

df = pd.DataFrame(columns = ['Наименование', 'Категория', 'Характеристики', 'Цена', 'Кол-во отзывов', 'Статус', 'Ссылка'])
soup = get_soup(main_url+df_category.loc[1, 'Ссылка'][1:])
category_number = 1
for p in range(int(soup.find_all('a', {'class': "pagination2__page"})[-1].text)):
    start = time.time()
    soup = get_soup(main_url+df_category.loc[category_number, 'Ссылка'][1:]+'?page='+str(p))
    
    try:
        r = soup.find_all('div', {'class':"catalog2__content-col"})
        for i, e in enumerate(r):
            try:
                get_url = get_info(e, 'a', 'img-link')['href']
                get_category = df_category.loc[category_number, 'Наименование']
                get_name = get_info(e, 'a', 'name').text
                reviews = get_info(e, 'a', 'rating-link')
                get_reviews_count = reviews.text.split(' ')[0] if reviews != None else 0
                status = [i for i in get_info(e, 'div', 'status').text.split('\n') if i != '']
                get_status = status[0] if len(status) != 0 else 'нет'
                price = get_info(e, 'span', 'price-val')
                get_price = int(''.join([i for i in price.text.split(' ') if i != ' '])) if price != None else 0
                chars = get_info(e, 'div', 'chars-body')
                get_chars = None if chars == None else dict(zip([i.text for i in get_info(e, 'div', 'chars-body').find_all('span', {'class':"name"})], 
                                                                [i.text for i in get_info(e, 'div', 'chars-body').find_all('span', {'class':"value"})])) 

                df = df.append({'Ссылка': get_url, 
                                'Категория': get_category, 
                                'Наименование': get_name, 
                                'Кол-во отзывов': get_reviews_count, 
                                'Статус': get_status, 
                                'Цена': get_price, 
                                'Характеристики': get_chars},
                               ignore_index = True)
            except Exception as e:
                print(p, i, e)
    except Exception as e:
        print(p, e)
    
    print(p, time.time()-start)
```
